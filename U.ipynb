<ol type="i">
    <li><h5>Sigmoid Aktivasyon Fonksiyon</h5>
     <ul type="square">
       <li>“S” şeklinde bir eğridir
       <li>Sigmoid Fonksiyon’nu kullanmamızın temel nedeni 0-1 arası çıktı vermesidir.Bu nedenle özelilikle tahmin etmemiz gereken modeller için kullanılır.Bunun anlamı sadece 0 ile 1 arasında çıktı aldığımız durumlarda kullanılması doğru bir seçimdir.
        <li>Türevlenebilirdir yani herhangi iki noktanın eğimini bulabiliriz.
    </ul>
  </li>
  
</ol>


```python
#import section
from matplotlib import pylab
import pylab as plt
import numpy as np

#sigmoid = lambda x: 1 / (1 + np.exp(-x))
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

mySamples = []
mySigmoid = []

# generate an Array with value ???
# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
# 
x = plt.linspace(-10,10,10)
y = plt.linspace(-10,10,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, sigmoid(x), 'r', label='linspace(-10,10,10)')
plt.plot(y, sigmoid(y), 'b', label='linspace(-10,10,100)')

# Draw the grid line in background.
plt.grid()

# Title & Subtitle
plt.title('Sigmoid Function')
plt.suptitle('Sigmoid')

# place the legen boc in bottom right of the graph
plt.legend(loc='lower right')

# write the Sigmoid formula
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 

# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
plt.show()
```

 <li> <h5>3.ReLU(Doğrultulmuş Duğrusal Ünite) Aktivasyon Fonksiyon</h5>
  
     <ul type="square">
       <li>ReLU fonksiyon en çok kullanılan fonksiyondur.Neredeyse tüm konvolüsyonel sinir ağlarında veya derin öğrenmede kullanılır. 
       <li>Grafikte görüldüğü gibi ReLU alttan yarım.Z sıfırdan küçük olduğunda f(z) sıfır ve z sıfırdan büyük veya eşit olduğunda f(z) z ye eşittir.
       <li>Tüm negatif değerlerin hemen sıfır olması  modelin verilere uygun şekilde uydurma veya bunları eğitme yeteneğini azaltacaktır.Bu da fonksiyona verilen herhangi bir negatif girdinin değeri grafikte hemen sıfıra döndüğü anlamına gelir.Yani negatif değerleri uygun şekilde eşleştirmeyerek sonuçtaki grafiği etkiler.
     </ul>
  </li>
  
</ol>

```python

from matplotlib import pylab
import pylab as plt
import numpy as np


def relu(x):
    return np.maximum(0, x)

mySamples = []
mySigmoid = []

# generate an Array with value ???
# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
# 
x = plt.linspace(-10,10,10)
y = plt.linspace(-10,10,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, relu(x), 'r', label='linspace(-10,10,10)')
plt.plot(y, relu(y), 'b', label='linspace(-10,10,100)')

# Draw the grid line in background.
plt.grid()

# Title & Subtitle
plt.title('ReLU Function')
plt.suptitle('ReLU')

# place the legen boc in bottom right of the graph
plt.legend(loc='lower right')

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 

# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
plt.show()
```
